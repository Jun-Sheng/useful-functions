```python
import csv
import re

word_file_name = "C:/Temp/words.txt"
    
with open(word_file_name, 'rb') as f:
    words = f.read().splitlines()

def split_matches(word):
    num_char = len(word)
    return_str = ""
    start_pos = 0
    end_pos = num_char
    for i in range(1, num_char+1):
        string = word[start_pos:i]
        #print str(start_pos) + "- " + str(i) + "-" + string
        
        if string in words:
            return_str = " ".join([return_str, string])
            start_pos = start_pos + len(string)
            #break
    
    return(return_str.lstrip())

def correctsentence (sentence):
    list_of_words = sentence.split()
    num_words = len(list_of_words)
    #print list_of_words

    #print num_words
    
    output_str = ""
    for i in range(0, num_words):
        word = list_of_words[i]
        #print "--> " + word
        
        if word in words:
            #print "Found " + word
            output_str = " ".join([output_str, word])
        else:
            output_str = " ".join([output_str, split_matches(word)])
            #print "Not Found " + split_matches(word)
            
    
    return(output_str.lstrip())

def main():

    #text = "TablePreferenceid3 and@ ip_userid"
    input_file_name = "C:/Temp/harmony.csv"  
    output_file_name = "C:/Temp/harmony_split.csv"  

    #word_file_name = "C:/Temp/words.txt"
    
    #with open(word_file_name, 'rb') as f:
    #    words = f.read().splitlines()

    output_file = open(output_file_name,'wb')
    writer = csv.writer(output_file)

    with open(input_file_name, 'rb') as inputcsvfile:
        input_file = csv.reader(inputcsvfile, delimiter=',')
        for row in input_file:
            text = row[1]
    
            # Split by Capital Letters
            text = " ".join(re.findall('[A-Z][^A-Z]*', text))
    
            # Replace _ / @ \ | with space
            text = text.replace("_", " ")
            text = text.replace("/", " ")
            text = text.replace("@", " ")
            text = text.replace("\\", " ")
            text = text.replace("|", " ")
    
            # convert text to lower case
            text = text.lower()
    
            # remove numbers
            text = ''.join([i for i in text if not i.isdigit()])
    
            # remove punctuation
            text = "".join(i for i in text if i not in ('.',',','!','?',':'))
    
            # strip whitespaces
            text = text.lstrip()
                        
            # Call "correctsentence" function defined above
            sent = correctsentence(text)
    
            # Split sentence per space
            sentence = sent.split(" ")

            if sentence != ['']:
                #print sentence
                writer.writerow(sentence)

    output_file.close()
    
if __name__ == "__main__":
    main()
```
